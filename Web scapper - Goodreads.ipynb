```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
URL = 'https://www.goodreads.com/quotes/tag/empowerment'
page = requests.get(URL)
```

```python
soup = BeautifulSoup(page.content,'html.parser')
```

```python
#soup
all_quote = []
```

```python
results = soup.find_all('div',class_='quoteText')
```

```python
for div in soup.find_all(class_='quoteText'):
    first_child = next(div.children, None)
    if first_child is not None:
        qut = first_child.string.strip() 
    author= div.find('span',class_='authorOrTitle')
    author = author.text.strip()
    all_quote.append({
        'author':author,
        'quote':qut
    })
```

```python
all_quote_df = pd.DataFrame(all_quote)
```

```python
all_quote_df.columns
```

```python
all_quote_df.author = [re.sub(r'[^\w\s]','',str_atr)  for str_atr in all_quote_df.author]
all_quote_df.quote = [re.sub(r'[^\w\s]','',str_qt)  for str_qt in all_quote_df.quote]
```
